```python
data2 = Support.get_data('ex1data2.txt')
data2.columns = ['Size in squre feet','No. of bedrooms','Price']
```

```python
normalised_data2 = data2.copy()
normalised_data2.columns = ['Normalised size', 'Normalised No. of bedrooms', 'Price']
normalised_data2['Normalised size'] = (data2['Size in squre feet'] - data2['Size in squre feet'].mean()) \
/ data2['Size in squre feet'].std()

normalised_data2['Normalised No. of bedrooms'] = (data2['No. of bedrooms'] - data2['No. of bedrooms'].mean()) \
/ data2['No. of bedrooms'].std()
```

```python
X2 = normalised_data2.loc[:,'Normalised size':'Normalised No. of bedrooms']
Y2 = normalised_data2['Price'].values
```

```python
predicted_y2 = Support.fit_GDregression(X2,Y2,n_iter=100)
```

```python
data2.loc[:,'Predicted price'] = predicted_y2
```
